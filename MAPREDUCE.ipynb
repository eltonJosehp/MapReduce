```python
%%writefile wordcount.py
from mrjob.job import MRJob
import sys

class WordCount(MRJob):
    def mapper(self, _, line):
        """Fase MAP: Divide líneas en palabras y emite (palabra, 1)."""
        for word in line.split():
            yield word.lower(), 1  # Convertimos a minúsculas

    def reducer(self, word, counts):
        """Fase REDUCE: Suma todas las ocurrencias de cada palabra."""
        yield word, sum(counts)

if __name__ == "__main__":
    sys.stdout.reconfigure(encoding='utf-8')  # Configurar la salida en UTF-8
    WordCount.run()
```
